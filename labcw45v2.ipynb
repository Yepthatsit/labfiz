```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
a = []
data = pd.read_excel('cw45.xlsx')
U = np.arange(150,250,25)
print(U)
R = 20
n = 154
print(data)
```

```python
r1 = 2
U2cm = [f"{i}V" for i in U]
X = np.square(list(data.loc[1,U2cm]))
Y = 2.48*10**12*U*R**2/(n**2*r1**2)
plt.scatter(X,Y)
model = sm.OLS(Y,X)
results = model.fit()
print(results.summary())
a.append(results.params[0])
plt.title(f"r = {r1}")
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(X,results.params[0]*X,c = 'r')
plt.show()
```
