```python
import pandas as pd
from scipy.stats import linregress
import numpy as np
final_results = []
data = pd.read_excel('labcw125.xlsx')
Ilog = np.log(data.loc[:,'Ipok[A]'])
results = linregress(data.loc[:,'Upok[V]'],Ilog)
print(f'{results.slope*(273.15 + 25.3)} +/- {results.stderr*(273.15 + 25.3)}')
final_results.append(results.slope*(273.15 + 25.3))
print(data.keys())
```

```python
Ilog1 = np.log(data.loc[:15,'I0st'])
results1 = linregress(data.loc[:15,'U0st'],Ilog1)
print(f'{results1.slope*(273.15)} +/- {results1.stderr*(273.15)}')
import matplotlib.pyplot as plt
#plt.title('Wykresy dla temperatury 0\xb0C')
#plt.subplot(1,2,1)
plt.ylabel('Ln(I)')
plt.xlabel('U')
plt.plot(data.loc[:15,'U0st'],results1.slope*data.loc[:15,'U0st'] + results1.intercept,c='r')
plt.scatter(data.loc[:15,'U0st'],Ilog1)
print(Ilog1)
final_results.append(results1.slope*(273.15))
#plt.subplot(1,2,2)
plt.figure()
plt.ylabel('I')
plt.xlabel('U')
plt.plot(data.loc[:15,'U0st'],data.loc[:15,'I0st'],marker = 'o')
plt.tight_layout()
plt.show()
```

```python
Ilog2 = np.log(data.loc[:15,'I9.5st'])
results2 = linregress(data.loc[:15,'U9.5st'],Ilog2)
print(f'{results2.slope*(273.15 +9.5)} +/- {results2.stderr*(273.15)}')
import matplotlib.pyplot as plt
#plt.subplot(1,2,1)
plt.ylabel('Ln(I)')
plt.xlabel('U')
plt.plot(data.loc[:15,'U9.5st'],results2.slope*data.loc[:15,'U9.5st'] + results2.intercept,c='r')
plt.scatter(data.loc[:15,'U9.5st'],Ilog2)
print(Ilog2)
final_results.append(results2.slope*(273.15 +9.5))
#plt.subplot(1,2,2)
plt.figure()
plt.plot(data.loc[:15,'U9.5st'],data.loc[:15,'I9.5st'],marker = 'o')
plt.ylabel('I')
plt.xlabel('U')
plt.tight_layout()
plt.show()
```
